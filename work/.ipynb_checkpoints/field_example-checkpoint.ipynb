```python
import numpy as np
import matplotlib.pyplot as plt

X, Y= np.meshgrid(np.arange(-2.2, 2.2, 0.4), np.arange(-2.2, 2.2, 0.4))
U = X/(np.sqrt(X**2+Y**2))
V = Y/(np.sqrt(X**2+Y**2))

plt.quiver(X, Y, U, V, color='red', angles='xy', scale_units='xy', scale=5.0)
# plt.xlim([-2,2])
# plt.ylim([-2,2])

plt.draw()
```
